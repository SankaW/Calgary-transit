```python
%sql
Select trip_id, count(stop_id) from calgary_transit.bronze.gtfs_stop_times where _ingest_date = '2026-02-23' group by trip_id order by count(stop_id) desc;
```

```python
%sql
Select st.* , t.route_id  from calgary_transit.bronze.gtfs_stop_times  as st
join  calgary_transit.bronze.gtfs_trips  as t
on  st.trip_id = t.trip_id
WHERE st._ingest_date = '2026-02-23'
  AND t._ingest_date = '2026-02-23';
```

```python
%sql
CREATE OR REPLACE VIEW calgary_transit.silver.v_stop_times_with_route AS
SELECT st.*, t.route_id
FROM calgary_transit.bronze.gtfs_stop_times st
JOIN calgary_transit.bronze.gtfs_trips t
  ON st.trip_id = t.trip_id
  WHERE st._ingest_date = '2026-02-23'
  AND t._ingest_date = '2026-02-23';
```

```python
%sql
Select route_id, count(stop_id) from calgary_transit.silver.v_stop_times_with_route where _ingest_date = '2026-02-23' group by route_id order by count(stop_id) desc;
```

```python
%sql
Select route_id, count(distinct stop_id) as stop_counts from calgary_transit.silver.v_stop_times_with_route where _ingest_date = '2026-02-23' group by route_id order by stop_counts desc;
```
