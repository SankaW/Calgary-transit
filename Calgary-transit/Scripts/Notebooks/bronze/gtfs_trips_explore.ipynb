```python
%sql
Select route_id,  count( trip_id) as count_route from calgary_transit.bronze.gtfs_trips where _ingest_date = '2026-02-18' group by  route_id order by count( trip_id) desc ;
```

```python
%sql
Select route_id, direction_id, count( trip_id) as count_route from calgary_transit.bronze.gtfs_trips where _ingest_date = '2026-02-18' group by  route_id, direction_id order by count( trip_id) desc;
```

```python
from pyspark.sql import functions as F

#load table
df_trips = spark.table('calgary_transit.bronze.gtfs_trips')

#filter by ingest date
filtered_df_trips = df_trips.filter(F.col('_ingest_date') == '2026-02-20')

#group and aggregate
grouped_df_trips = filtered_df_trips.groupBy('route_id').agg(F.count('trip_id').alias('trip_count')).orderBy(F.desc('trip_count'))

display(grouped_df_trips)
```
