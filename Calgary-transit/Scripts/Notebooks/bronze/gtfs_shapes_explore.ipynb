```python
%sql
select count(*) from calgary_transit.bronze.gtfs_shapes where `_ingest_date` = '2026-02-23' ;
```

```python
%sql
select count(*) from calgary_transit.bronze.gtfs_shapes where `_ingest_date` = '2026-02-23';
```

```python
%sql
select _ingest_date, count(*) from calgary_transit.bronze.gtfs_shapes group by `_ingest_date` ;
```

```python
%sql
select shape_id, shape_pt_lat, shape_pt_lon, shape_pt_sequence, shape_dist_traveled  from calgary_transit.bronze.gtfs_shapes where `_ingest_date` = '2026-02-23';
```

```python
%sql
select distinct shape_id from calgary_transit.bronze.gtfs_shapes where `_ingest_date` = '2026-02-23';
```

```python
%sql
select  shape_id , count(shape_id) as points from calgary_transit.bronze.gtfs_shapes where `_ingest_date` = '2026-02-23' group by shape_id order by points desc;
```
