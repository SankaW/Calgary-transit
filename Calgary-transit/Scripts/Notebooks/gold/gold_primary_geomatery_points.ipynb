## Primary geometry points (join to shapes + downsample)

```python
%sql
CREATE OR REPLACE TABLE calgary_transit.gold.route_primary_geometry_points_full AS
SELECT
  p.route_id,
  p.route_short_name,
  p.route_long_name,
  p.route_type,
  p.route_color,
  p.route_text_color,
  p.direction_id,
  p.primary_headsign,
  p.primary_shape_id AS shape_id,
  p.primary_trip_count AS trip_count,
  sh.shape_pt_lat,
  sh.shape_pt_lon,
  sh.shape_pt_sequence
FROM calgary_transit.gold.route_primary_shape p
JOIN calgary_transit.bronze.gtfs_shapes sh
  ON sh.shape_id = p.primary_shape_id
 AND sh._ingest_date = '2026-02-23';
```

```python
%sql
select Count(*) from calgary_transit.gold.route_primary_geometry_points_full ;
```

```python
%sql
select * from calgary_transit.gold.route_primary_geometry_points_full ;
```

```python
%sql
CREATE OR REPLACE TABLE calgary_transit.gold.route_primary_geometry_points_lite AS
SELECT
  p.route_id,
  p.route_short_name,
  p.route_long_name,
  p.route_type,
  p.route_color,
  p.route_text_color,
  p.direction_id,
  p.primary_headsign,
  p.primary_shape_id AS shape_id,
  sh.shape_pt_lat,
  sh.shape_pt_lon,
  sh.shape_pt_sequence
FROM calgary_transit.gold.route_primary_shape p
JOIN calgary_transit.bronze.gtfs_shapes sh
  ON sh.shape_id = p.primary_shape_id
 AND sh._ingest_date = '2026-02-23'
WHERE (sh.shape_pt_sequence % 5) = 0
   OR sh.shape_pt_sequence = 1;
```

```python
%sql
select Count(*) from calgary_transit.gold.route_primary_geometry_points_lite ;
```
