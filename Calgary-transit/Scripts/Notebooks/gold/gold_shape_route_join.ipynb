## Primary shape table (rank = 1) + join routes for labels/colors

```python
%sql
CREATE OR REPLACE TABLE calgary_transit.gold.route_primary_shape AS
SELECT
  r.route_id,
  r.route_short_name,
  r.route_long_name,
  r.route_type,
  r.route_color,
  r.route_text_color,
  s.direction_id,
  s.shape_id AS primary_shape_id,
  s.dominant_headsign AS primary_headsign,
  s.trip_count AS primary_trip_count,
  s.total_trips_dir,
  s.dominance_ratio,
  CASE WHEN s.dominance_ratio < 0.70 THEN true ELSE false END AS is_multi_variant
FROM calgary_transit.gold.route_shape_ranking s
JOIN calgary_transit.bronze.gtfs_routes r
  ON r.route_id = s.route_id
 AND r._ingest_date = '2026-02-23'
WHERE s.shape_rank = 1;
```
