## Route–Stop set (distinct stops per route, deduped)

```python
%sql
CREATE OR REPLACE TABLE calgary_transit.gold.route_stopset AS
WITH trips_f AS (
  SELECT trip_id, route_id, direction_id
  FROM calgary_transit.bronze.gtfs_trips
  WHERE _ingest_date = '2026-02-23'
),
route_stop AS (
  SELECT DISTINCT
    t.route_id,
    t.direction_id,
    st.stop_id
  FROM trips_f t
  JOIN calgary_transit.bronze.gtfs_stop_times st
    ON st.trip_id = t.trip_id
   AND st._ingest_date = '2026-02-23'
)
SELECT
  rs.route_id,
  r.route_short_name,
  r.route_long_name,
  rs.direction_id,
  rs.stop_id,
  s.stop_name,
  s.stop_lat,
  s.stop_lon
FROM route_stop rs
JOIN calgary_transit.bronze.gtfs_routes r
  ON r.route_id = rs.route_id
 AND r._ingest_date = '2026-02-23'
JOIN calgary_transit.bronze.gtfs_stops s
  ON s.stop_id = rs.stop_id
 AND s._ingest_date = '2026-02-23';
```

```python
%sql
select * from calgary_transit.gold.route_stopset;
```
