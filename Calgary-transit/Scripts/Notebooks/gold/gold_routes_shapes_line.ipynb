```python
%sql
-- distinct shape_ids and number of point each of them has

select  shape_id , count(shape_id) as points from calgary_transit.bronze.gtfs_shapes where `_ingest_date` = '2026-02-23' group by shape_id order by points desc; 
```

Here we have 676 shape ids , now I wanna connect this ids with route id.

```python
%sql
--  shapeids against routeid 
Select route_id,  count( distinct shape_id) as count_shape from calgary_transit.bronze.gtfs_trips where _ingest_date = '2026-02-23' group by  route_id 
order by count( distinct shape_id)  desc ;
```

Accoding to above results, majority route_id can have more than one shape ids.

```python
%sql
Select distinct route_id, shape_id  from calgary_transit.bronze.gtfs_trips where _ingest_date = '2026-02-23'  and route_id = '201-20765'
```

Based on route id = "201-20765" let's try to find out what are the other feild that may cause to multiple shapes

```python
%sql

Select distinct route_id, direction_id, shape_id  from calgary_transit.bronze.gtfs_trips where _ingest_date = '2026-02-23'  and route_id = '201-20765'
```

```python
%sql
--   routeids against shapeid 
Select  shape_id,  count( distinct route_id) as count_routes from calgary_transit.bronze.gtfs_trips where _ingest_date = '2026-02-23' group by  shape_id 
order by count( distinct route_id)  desc ;
```

According to above query only 2 space id has two (multiple) route every other one just have one routeid.
