Delete for ALL  GTFS tables in one run

```python
dbutils.widgets.text("catalog", "calgary_transit")
dbutils.widgets.text("schema", "bronze")
dbutils.widgets.text("ingest_date", "")  # 2026-02-17

catalog = dbutils.widgets.get("catalog").strip()
schema  = dbutils.widgets.get("schema").strip()
d       = dbutils.widgets.get("ingest_date").strip()

tables = [
  "gtfs_agency","gtfs_calendar","gtfs_calendar_dates","gtfs_routes",
  "gtfs_shapes","gtfs_stop_times","gtfs_stops","gtfs_trips"
]

for t in tables:
    full = f"{catalog}.{schema}.{t}"
    spark.sql(f"DELETE FROM {full} WHERE _ingest_date = '{d}'")
    print("deleted", full, d)
```
